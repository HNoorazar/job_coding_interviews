### Minimum Size Subarray Sum

Given an array of ```n``` positive integers and a positive integer ```s```, find the minimal length of a contiguous subarray of which the ```sum ≥ s```.

````python
def min_subarray_sum(arr, s):
    
    arr_len = len(arr)
    minimum_length = arr_len + 1
    
    if arr_len==0:
        print ("arrray is empty!")
        return 0
    
    if max(arr)>s:
        return 1
    
    for start_idx in range(len(arr)-1):
        accum = arr[start_idx]
        far_right = start_idx
        while (accum < s and (far_right < len(arr)-1) ):
            far_right += 1
            accum += arr[far_right]
            if accum >= s:
                minimum_length = min(minimum_length, far_right - start_idx+1)

    if minimum_length < arr_len:
        return minimum_length
    else:
        return 0
````

### number of Subarrays that Sum up to  K
(Subarray Sum Equals ```K```)

Given an array of integers and an integer ```k```,
you need to find the total number of 
continuous subarrays whose sum equals to ```k```.

````python
def subarray_sum_count(arr, k):
    """
      input: arr is an array of integers (positive numbers)
             k: an integer
      
      output: count of subarrays that add up to k
    """
    arr_len = len(arr)
    
    if arr_len==0:
        print ("array is empty")
        return
    
    count_of_subarray = 0
    left_pointer = 0
    
    while left_pointer < arr_len - 1:

        # if k is in the list, count it as an subarray
        # and move to the right.
        if arr[left_pointer] == k:
            count_of_subarray += 1
            left_pointer += 1
        
        # if current element is greater than k
        # move to the right. Adding another 
        # positive number would not help!
        elif arr[left_pointer] > k :
            left_pointer += 1
        
        elif arr[left_pointer] < k:
            acc_sum = arr[left_pointer]
            far_right = left_pointer
            while acc_sum < k:
                far_right += 1
                acc_sum += arr[far_right]
                if acc_sum>k:
                    left_pointer += 1
                    
                elif acc_sum==k:
                    count_of_subarray += 1
                    left_pointer += 1
    if arr[-1] == k:
        count_of_subarray += 1
        
    return count_of_subarray
````

````python
def merge_intervals(intervals):
    intervals.sort(key=lambda x: x.start)

    merged = []
    for interval in intervals:
        # if the list of merged intervals is empty or if the current
        # interval does not overlap with the previous, simply append it.
        if not merged or merged[-1].end < interval.start:
            merged.append(interval)
        else:
        # otherwise, there is overlap, so we merge the current and previous
        # intervals.
            merged[-1].end = max(merged[-1].end, interval.end)
    return merged
````

### Search in Rotated Sorted Array

Suppose an array sorted in ascending order is rotated at some pivot unknown to you beforehand. (i.e., ```[0,1,2,4,5,6,7]``` might become ```[4,5,6,7,0,1,2]```).

You are given a target value to search. If found in the array return its index, otherwise return -1. You may assume no duplicate exists in the array. Your algorithm's runtime complexity must be in the order of O(log n).

**Example 1:**
```python
Input: nums = [4,5,6,7,0,1,2], target = 0
Output: 4
```

**Example 2:**
```python
Input: nums = [4,5,6,7,0,1,2], target = 3
Output: -1
```

````python
def find_query_sorted_array(arr, query):
    max_idx = find_max_idx_via_min(arr)
    
    if query>= arr[0] and query<= arr[max_idx]:
        result = binary_search(arr[0:max_idx+1], query)
        return result
    else:
        result = binary_search(arr[max_idx+1:], query)
        if type(result) == int:
            return result + max_idx + 1

def find_max_idx_via_min(arr):
    left = 0
    right = len(arr) - 1
    while left<right:
        if arr[left] < arr[right]:
            if left==0:
                return len(arr)-1
            else:
                return left-1
            
        mid = int((right+left)/2)
        if arr[mid] >= arr[left]:
            left = mid + 1
        else:
            right = mid
    
    if left ==0:
        return len(arr)-1
    else:
        return left-1

def binary_search(arr, query):
    """
      arr: a given list we want to find an element in
      query: a given number we want to find in arr
    """
    # the index of left boundary of sub-list
    left = 0
    # the index of right boundary of sub-list
    right = len(arr)-1
    
    while (left <= right):
        mid = int((left+right)/2)
        
        if query==arr[mid]:
            return mid
        
        elif query<arr[mid]:
            right = mid-1
            
        elif query>arr[mid]:
            left = mid+1

    print ('**Bad News** The query is not in the list!!!')
    return 'not_found'
````

### Meeting Rooms

````python
def min_meeting_rooms(intervals):
    """
    Assume intervals is list of lists
    """
    
    no_meetings = len(intervals)
    start_times = [0]*no_meetings
    end_times = [0]*no_meetings
    
    # put starting/ending times of meetings in the proper list 
    # O(N) where N is number of meetings.
    for ii in range(no_meetings):
        start_times[ii] = intervals[ii][0]
        end_times[ii] = intervals[ii][1]

    # sort starting/ending times O(NLogN)
    start_times.sort()
    end_times.sort()
    
    
    start_pointer = 0
    end_pointer = 0
    count_rooms = 0
    min_rooms = 0
    
    while start_pointer < no_meetings:
        if start_times[start_pointer] < end_times[end_pointer]:
            count_rooms += 1
            min_rooms = max(min_rooms, count_rooms)
            start_pointer += 1
        else:
            count_rooms -= 1
            end_pointer += 1
    return min_rooms
min_meeting_rooms([[0,30], [10, 15], [14, 24]])
````

### Intersection of Two Linked Lists

A:          a1 → a2

                   ↘
                     c1 → c2 → c3 
                   ↗            
B:     b1 → b2 → b3

````python
def linked_list_intersection(head1, head2):
    """
    Two pointers, one for each list.
    traverse the lists one at a time till 
    one of the lists hits the end. Then start
    counting how many more elements is left on
    the unfinished list.
    
    then start the process again, however, 
    this time, let the longer list takes the 
    extra steps at first, and then start evolving the 
    second pointer.
    
    Carry on, till both pointers reach the intersection 
    point.
    
    This way we are essentially go for finding the
    intersection start when both lists have the same length!
    It is like shifting the lonegr list to the left so
    both lists have their end at the same spot.
    """
    lenght_1 = get_list_length(head1)
    lenght_2 = get_list_length(head2)
    
    if lenght_1==0 or lenght_2==0:
        print ("There is no intersection. At least one list is empty.")
        return 0
    
    if lenght_1 > lenght_2:
        extra_element_count = lenght_1 - lenght_2
        pointer_2 = head2
        pointer_1 = head1
        for ii in range(extra_element_count):
            pointer_1 = pointer_1.next
        
    elif lenght_2 > lenght_1:
        extra_element_count = lenght2 - lenght1
        pointer_1 = head1
        pointer_2 = head2
        for ii in range(extra_element_count):
            pointer_2 = pointer_2.next
    
    elif lenght_2 == lenght_1:
        pointer_1 = head1
        pointer_2 = head2
    
    while pointer_1.val != pointer_2.val:
        pointer_1 = pointer_1.next
        pointer_2 = pointer_2.next
        
    return pointer_1
  

def get_list_length(head):
    if head.val == None:
        return 0
    
    elif head.val!= None and head.next==None:
        return 1
    
    else:
        length = 1
        current = head
        while current.next != None:
            length += 1
            current = current.next
        
        return length
````

#### [Merge Two Sorted (linked) Lists](https://leetcode.com/problems/merge-two-sorted-lists/)

```python
Input: 1->2->4, 1->3->4
Output: 1->1->2->3->4->4
```

````python
def merge_two_lists(l1, l2):
    """
    Merge two sorted linked lists
    """
    # initiate the merges linked list
    merged_list_head = list_node(0)
    
    # current node, i.e. a pointer to the current location
    current_node = merged_list_head
    
    while True:
        if l1 == None:
            current_node.next = l2
            break
        if l2 == None:
            current_node.next = l1
            break
        if l1.val < l2.val:
            current_node.next = l1
            current_node = current_node.next
            l1  = l1.next
        else:
            current_node.next = l2
            current_node = current_node.next
            l2 = l2.next
    return merged_list_head.next
````

````python
def add_two_numbers(head1, head2):
    """
      input: head1 and head2 are heads of two linked list
             we want to add two numbers represented by these two 
             linked list, where they are non empty, and each node 
             in the list represents a digit betweein 0 and 9.
             A list represents digits in reverse order of the number.
             
      output: a linked list which represents the sum of input lists.
    """
    # check if head1 and head2 are actually nodes
    # or empty lists.
    if type(head1) == list and type(head2) == list:
        print ("Inputs are not of type linked list")
        return
    
    if type(head1)==list and type(head2)!= list:
        return head2
    
    if type(head2) == list and type(head1) != list:
        return head1

    current1 = head1
    current2 = head2
    carry_on = 0
    
    sum_head = list_node(0)
    current_sum_Node = sum_head
    
    while (current1.next != None) or (current2.next != None):
        if (current1.val == None):
            current1_value = 0
        else:
            current1_value = current1.val
        
        if current2.val == None:
            current2_value = 0
        else:
            current2_value = current2.val
            
        summation = current1_value + current2_value + carry_on
        carry_on = int(summation / 10)
        current_sum_Node.val = summation % 10
        current_sum_Node.next = list_node(0)
        current_sum_Node = current_sum_Node.next
        
        if (current1.next != None) and (current2.next != None):
            current1 = current1.next
            current2 = current2.next

        #  if list 2 is shorter
        elif (current1.next != None) and (current2.next == None):
            current1 = current1.next
            current2 = list_node(0)

        # if list 1 is shorter    
        elif (current1.next == None) and (current2.next != None):
            current1 = list_node(0)
            current2 = current2.next
    
    # the last step where last nodes both points are pointing to None
    summation = current1.val + current2.val + carry_on
    carry_on = int(summation / 10)
    current_sum_Node.val = summation % 10
    if carry_on==1:
        current_sum_Node.next = list_node(1)

    return sum_head
````

###  [Smallest Range](https://leetcode.com/problems/smallest-range/description/)

You have ```k``` lists of sorted integers in ascending order. Find the smallest range that includes at least one number from each of the `k` lists.

We define the range `[a,b]` is smaller than range `[c,d]` if `b-a < d-c` or `a < c` if `b-a == d-c`.


**Input**:```[[4,10,15,24,26], [0,9,12,20], [5,18,22,30]]```

**Output**: ```[20,24]```

**Explanation**:

List 1: ```[4, 10, 15, 24, 26]```, 24 is in range ```[20,24]```.

List 2: ```[0, 9, 12, 20]```, 20 is in range ```[20,24]```.

List 3: ```[5, 18, 22, 30]```, 22 is in range ```[20,24]```.

````python
def find_smallest_range(list_of_lists):
    """
     input: list_of_lists is a list whose elements 
            are lists.
            
     output: the shortest range that contains 
             at least one element of each entry 
             of the input.
    """
    # length of the input
    length = len(list_of_lists)
    
    # keep the shortest range so far in shortest_range
    shortest_range = float('inf')
    
    # save current elements being under
    # consideration in the following list.
    current_entries = [None] * length
    for ii in range(length):
        if len(list_of_lists[ii]) == 0:
            print ("The {}th element of the list is empty.".format(ii))
            return 
        current_entries[ii] = list_of_lists[ii][0]

    current_pointers = [0] * length
    not_exahusted = True
    
    while not_exahusted:
        """
         min_idx is the index of the minimum element 
         in current_entries which is index of the 
         list in list_of_lists. So, it points to 
         the list in list_of_lists containing the 
         current minimum element.
        """
        min_idx = current_entries.index(min(current_entries))
        
        # find current range
        current_range = max(current_entries) - current_entries[min_idx]
        
        # replace shortest range if current range 
        # is smaller than the range in hand.
        if current_range <= shortest_range:
            shortest_range = current_range

        if current_pointers[min_idx] <= (len(list_of_lists[min_idx]) - 2):
            current_pointers[min_idx] += 1
            current_entries[min_idx] = list_of_lists[min_idx][current_pointers[min_idx]]
        else:
            not_exahusted = False
                
    return shortest_range
````

### Valid Palindrome

**Palindrome**: a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.

Given a string, determine if it is a palindrome, considering only alphanumeric characters and ignoring cases.

Note: For the purpose of this problem, we define empty string as valid palindrome.

------------------
**Example 1**

**Input**: "A man, a plan, a canal: Panama"

**Output**: true

------------------
**Example 2**

**Input**: "race a car"

**Output**: false


````python
def valid_palindrome(string):
    """
     input : string is a given phrase
    """
    str_len = len(string)
    left_pointer = 0
    right_pointer = str_len - 1
    flag = True
    
    if str_len==0 or str_len==1:
        return True
    else:
        while left_pointer < right_pointer and flag:
            while not(string[left_pointer].isalnum()):
                left_pointer += 1

            while not(string[right_pointer].isalnum()):
                right_pointer -= 1

            # the following if-clause is just for security check!
            if string[left_pointer].isalnum() and string[right_pointer].isalnum():
                if string[left_pointer].lower() != string[right_pointer].lower():
                    flag = False
                    return False
                else:
                    left_pointer += 1
                    right_pointer -= 1                
    return True
````

Judge whether you can make it a palindrome.

**Example 2:**

Explanation: You could delete the character ```c```.

````python
def valid_palindrome_II(string):
    """
    input: string
    
    We create a dictionary of count of letters.
    If number of characters with odd numbers is more than 1,
    then is not doable.
    """
    str_len = len(string)
    
    if str_len<=2:
        return True
    
    
    # count number of each character in the sting:
    char_count = {}
    for character in string:
        if character in char_count:
            char_count[character] += 1
        else:
            char_count[character] = 1
            
    # count number of characters that are repeated odd number of times.
    
    repeated_once = 0 # if there is one of these, we can delete it.
    repeated_odd_more_than_once = 0
    
    for key in char_count:
        if char_count[key]==1:
            repeated_once += 1
        elif ((char_count[key] % 2) != 0):
            repeated_odd_more_than_once += 1

    if repeated_odd_more_than_once>1:
        return False

    elif (repeated_once<=2) and (repeated_odd_more_than_once==0):
        return True

    elif (repeated_once==1) and (repeated_odd_more_than_once==1):
        return True
   
    else:
        return False
valid_palindrome_II('abbbdccca')
````

### [Valid Parentheses](https://leetcode.com/problems/valid-parentheses/)
Given a string containing just the characters ```'('```, ```')'```, ```'{'```, ```'}'```, ```'['``` and ```']'```, determine if the input string is valid.

An input string is valid if:

  1. Open brackets must be closed by the same type of brackets.
  2. Open brackets must be closed in the correct order.

Note that an empty string is also considered valid.

---------------
** Example **

**Input**: "()[]{}"

**Output**: true

---------------

````python
def valid_parantheses(string):
    """
     My solution is of O(n). 
     Once we go through all the elements of string
     and save the open paratheses type in one list,
     and closing parathesis types in another list,
     in the order of their appearance in strin. And 
     then check them one by one.
    """
    opening = ['(', '[', '{']
    closing = [')', ']', '}']
    string_len = len(string)
    if string_len==0:
        return True

    # if there are odd number of paratheses
    # then, of course, it cannot be a valid case.
    if string_len % 2 != 0 or (string[-1] in opening) or (string[0] in closing):
        return False
    
    status = []
    
    for ii in range(string_len):
        if string[ii] in opening:
            status.append(string[ii])
            
        else:
            if string[ii] == ')' and status[-1]=='(':
                status.pop()
            elif string[ii] == ']' and status[-1]=='[': 
                status.pop()
            elif string[ii] == '}' and status[-1]=='{':
                status.pop()
            else:
                return False
    if len(status)==0:
        return True
````
